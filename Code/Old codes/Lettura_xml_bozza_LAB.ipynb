```python
list_event=[]
len(root)
```

```python
for i in range(2,len(root)):
  #print(i)
  for child in root[i]:
    if((child.text)!= None):  
      string_event=child.text 
      event=np.array([int(s) for s in string_event.split() if s.isdigit()])
      list_event.append(event)
```
